## Linear Programming Using Simplex Method
### ReVelle et al. book, Ch. 4, Ex. 4.3 p.113

```python
# Import python packages numpy, matplotlib, and pulp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pulp
```

### Problem Statement
Solve  the  following  linear  program  using  the  simplex  method. Compute the value of the objective function and decision variables at optimality, and indicate which statement best describes the solution and why:

### Objective Function & Constraints
#### Maximizing
Maximize $Z=3x_1+2x_2$            &ensp;&thinsp;&ensp;&thinsp;  (a) this linear program has a unique optimal solution

#### costraints
Subject to:
> $-x_1+2x_2\leq 6$               &ensp;&thinsp;&ensp;&thinsp;&ensp;&thinsp;    (b) this linear program has alternate optima\
> $2x_1-5x_2\leq 10$              &ensp;&thinsp;&ensp;&thinsp;&ensp;&thinsp;    (c) this linear program is infeasible \
> $-2x_1+2x_2\leq 2$              &ensp;&thinsp;&ensp;&thinsp;&ensp;            (d) this linear program is unbounded \
> $x_1,x_2\geq 0$

## Model Formulation
### Decision Variables
$x_1$: the first variable \
$x_2$: the second variable

### Problem Solution
#### Graphical Solution

```python
# Construct lines
# x1, x2 > 0
x = np.linspace(0, 100, 2000)
y = np.linspace(0, 100, 2000)

# -x1+2x2 <= 6
y1 = (6+x)/2.0

# 2x1-5x2 <= 10
y2 = (10-2*x)/(-5.0)

# -2x1+2x2 <= 2
y3 = (2+2*x)/(2.0)
```

Then, we plot the constraints as equalities:

```python
# Make plot
plt.plot(x, y1, label=r'$-x_1+2x_2 \leq 6$')
plt.plot(x, y2, label=r'$2x_1-5x_2\leq 10$')
plt.plot(x, y3, label=r'$-2x_1+2x_2\leq 2$')
plt.xlim((0, 20))
plt.ylim((0, 20))
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Then, we fill the feasible region in accordance to the constrains inequalities:

```python
# Make plot
plt.plot(x, y1, label=r'$-x_1+2x_2 \leq 6$')
plt.plot(x, y2, label=r'$2x_1-5x_2\leq 10$')
plt.plot(x, y3, label=r'$-2x_1+2x_2\leq 2$')
plt.xlim((0, 20))
plt.ylim((0, 20))
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
# Fill feasible region
y4 = np.minimum(y1,y3)
plt.fill_between(x, y4, y2,  color='grey', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Our solution lies somewhere in the grey feasible region in the graph above, and as you see, the feasible region is unbounded.\
\
It has been proven that the minima and maxima of linear programming problems lie at the vertices of the feasible region. In this example, there are only 4 corners to our feasible region, so we can find the solutions for each corner to find our maximum. 

### Solution
We need to examine the value of the objective function $Z$ at the 4 corners, which are the intersections of the system of equations below. 

| Point | $x_1$ | $x_2$ | $s_1$ | $s_2$ | $s_3$ | $Z$ Value |
|-----|--------|--------|---------|---------|---------|---------|
| $A$ | $0.00$ | $0.00$ | $6.00$  | $10.00$ | $2.00$  | $0.00$  |
| $B$ | $5.00$ | $0.00$ | $11.00$ | $0.00$  | $12.00$ | $15.00$ |
| $C$ | $4.00$ | $5.00$ | $0.00$  | $27.00$ | $0.00$  | $22.00$ |
| $D$ | $0.00$ | $1.00$ | $4.00$  | $15.00$ | $0.00$  | $2.00$  |

The **maximum solution** $Z=22.00$ is for $x_1=4$ and $x_2=5$. However, since the feasible region is unbounded, we shoudl check how $Z$ behaves on the unbounded side of the region.
Below we plot the feasible region with the objective function and we see that $3x_1+2x_2=22$ intersects with the feasible region at $x_1=4$ and $x_2=5$. However, parallel lines to $3x_1+2x_2=22$ exist on the unbounded side with higher $Z$ value. So, the solution is unbounded as well.

```python
# Objective Function Value 1
y5 = (22-3*x)/2.0
# Objective Function Value 2
y6 = (32-3*x)/2.0
# Objective Function Value 3
y7 = (12-3*x)/2.0
# Make plot
plt.plot(x, y1, label=r'$-x_1+2x_2 \leq 6$')
plt.plot(x, y2, label=r'$2x_1-5x_2\leq 10$')
plt.plot(x, y3, label=r'$-2x_1+2x_2\leq 2$')

plt.plot(x, y5, 'k--', label='$3x_1+2x_2=22$')
plt.plot(x, y6,'k-.', label='$3x_1+2x_2=32$')
plt.plot(x, y7, 'k',label='$3x_1+2x_2=12$')

plt.xlim((0, 20))
plt.ylim((0, 20))
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
# Fill feasible region
y4 = np.minimum(y1,y3)
plt.fill_between(x, y4, y2,  color='grey', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)



```
